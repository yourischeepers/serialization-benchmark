# Charts generator

This file uses Kotlin Notebooks and the Kandy library to generate charts that analyze and visualize the data from the benchmarks.

```kotlin
%useLatestDescriptors
%use dataframe
%use kandy(0.8.3)

import org.jetbrains.kotlinx.dataframe.api.*
import org.jetbrains.kotlinx.dataframe.io.readCSV

val raw = DataFrame.readCsv("results.csv")

val df = raw
    .rename("Param: complexity").into("Complexity")
    .rename("Param: size").into("Size")
    .rename("Score Error (99,9%)").into("ErrorMargin")
    .convert("Score")
    .with { it.toString().replace(",", ".").toDouble() }
    .convert("ErrorMargin")
    .with { it.toString().replace(",", ".").toDouble() }
    .convert("Size")
    .with { it.toString().toInt() }
    .convert("Benchmark")
    .with { if (it.toString().endsWith("gsonDeserialize")) "Gson" else "Kotlinx" }
    .select("Benchmark", "Complexity", "Size", "Score", "ErrorMargin", "Unit")
    .add("MicrosecondsPerOp") { 1_000_000.0 / ("Score"<Double>()) }


df
```

```kotlin
val complexDf = df.filter {
    Complexity == "complex"
}

complexDf.plot {
    line {
        x(Size)
        y(MicrosecondsPerOp)
        color(Benchmark)
    }

    points {
        x(Size)
        y(MicrosecondsPerOp)
        color(Benchmark)
    }

    layout {
        title = "Deserialization time (complex payloads)"
        xAxisLabel = "Object count"
        yAxisLabel = "Microseconds per operation"
    }
}
```

```kotlin
complexDf.plot {
    line {
        x(Size)
        y(Score)
        color(Benchmark)
    }

    points {
        x(Size)
        y(Score)
        color(Benchmark)
    }

    layout {
        title = "Operations count (complex payloads)"
        xAxisLabel = "Object count"
        yAxisLabel = "Operations per second"
    }
}
```

```kotlin
val complexBarDf = complexDf
    .sortBy("Size")
    .add("Label") {
        "${Size} objects - ${Benchmark}"
    }
    .add("ErrorMin") {
        ("Score"<Double>()) - ("ErrorMargin"<Double>())
    }
    .add("ErrorMax") {
        ("Score"<Double>()) + ("ErrorMargin"<Double>())
    }

complexBarDf.plot {
    bars {
        x("Label") {
            axis.name = ""
        }
        y("Score")
        fillColor(Benchmark)
    }

    errorBars {
        x("Label")
        y("Score")
        yMin("ErrorMin")
        yMax("ErrorMax")
    }

    coordinatesTransformation =
        CoordinatesTransformation.cartesianFlipped()

    layout {
        title = "Complex payload throughput"
        yAxisLabel = "Operations per second"

        size = 900 to 500
    }
}
```

```kotlin
val complexBarDf = complexDf
    .sortBy("Size")
    .add("Label") {
        "${Size} objects - ${Benchmark}"
    }
    .add("ErrorMin") {
        val score = "Score"<Double>()
        val margin = "ErrorMargin"<Double>()
        1_000_000.0 / (score + margin)
    }
    .add("ErrorMax") {
        val score = "Score"<Double>()
        val margin = "ErrorMargin"<Double>()
        1_000_000.0 / (score - margin)
    }

complexBarDf.plot {
    bars {
        x("Label") {
            axis.name = ""
        }
        y("MicrosecondsPerOp")
        fillColor(Benchmark)
    }

    errorBars {
        x("Label")
        y("MicrosecondsPerOp")
        yMin("ErrorMin")
        yMax("ErrorMax")
    }

    coordinatesTransformation =
        CoordinatesTransformation.cartesianFlipped()

    layout {
        title = "Complex payload deserialization time"
        yAxisLabel = "Microseconds per operation"
        size = 900 to 500
    }
}
```
